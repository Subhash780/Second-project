```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_df=pd.DataFrame(scaler.fit_transform(df.iloc[:,1:]))
scaled_df.head()
```

```python
scaled_df.info()
```

```python
import warnings
warnings.filterwarnings("ignore")
wcss=[]
for i in range(1,10):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(scaled_df)
  wcss.append(kmeans.inertia_)
print(wcss)
plt.plot(range(1, 10), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
clusters_new=KMeans(3)
clusters_new.fit(scaled_df)
```

```python
clusters_new.labels_
```

```python
set(clusters_new.labels_)
```

```python
df['clusterid_new']=clusters_new.labels_
df.head()
```

```python
df.sort_values(by="clusterid_new")
```

```python
clusters_new.cluster_centers_
```

```python
df[df['clusterid_new']==0]
```

```python
df[df['clusterid_new']==1]
```

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(scaled_df,clusters_new.labels_,metric='euclidean')
score
```

```python
df1=pd.DataFrame()
df1['cluster_id']=clusters_new.labels_
df1.head()
```

```python

from sklearn.cluster import DBSCAN
```

```python
df=pd.read_csv("C:/Users/chopp/Downloads/crime_data.csv")
df.head()
```

```python

df=df.rename(columns={"Unnamed: 0":"States"})
```

```python
df.head()
```

```python
df.info()
```

```python
array=df.values
array
```

```python
stscaler=StandardScaler()
X=stscaler.fit_transform(array)
X
```

```python
from sklearn.neighbors import NearestNeighbors

nearest_neighbors = NearestNeighbors(n_neighbors=5)
neighbors = nearest_neighbors.fit(X)

distances, indices = neighbors.kneighbors(X)
print(distances)
# print(indices)
# print(distances.shape)
distances = np.sort(distances[:,4], axis=0)
#print(distances)
fig = plt.figure(figsize=(5, 5))
plt.plot(distances)
plt.xlabel("Points")
plt.ylabel("Distance")
plt.grid()
```

```python
dbscan=DBSCAN(eps=1.3,min_samples=5)
clusters=dbscan.fit_predict(X)
```

```python
dbscan.labels_
```

```python
df['cluster']=dbscan.labels_
df
```

```python
from sklearn import cluster
df[df['cluster']==-1]
```

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(X,dbscan.labels_,metric='euclidean')
score
```
